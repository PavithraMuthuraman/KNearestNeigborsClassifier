**Step 1: Data Collection**

```python
#Import necessary libraries
import numpy as np
import pandas as pd
```

```python
#Load the dataset
data=pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
data.head()
```

**Step 2: Data Cleaning**

```python
df=data.drop(['User ID'],axis=1)
```

```python
df['Gender']=df['Gender'].apply({'Male':1,'Female':2}.get)
```

```python
df.head()
```

**Step 3 : Dividing the data into dependent and independent variables**

```python
x=df[['Gender','Age','EstimatedSalary']] #independent variables
y=df['Purchased'] #dependent variables
```

**Step 4 : Splitting the data into training and testing set**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,train_size=0.85)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Initialize the K-Nearest Neighbors classifier with 5 neighbors
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
#Train the model
knn.fit(x_train,y_train)
```

```python
#Make predictions on test data
predictions=knn.predict(x_test)
```

**Adding new customer data to predict product purchase**

```python
new_data={'Gender':1,'Age':30,'EstimatedSalary':80000}
index=[0]
```

```python
# Convert the new data into a DataFrame
my_data=pd.DataFrame(new_data,index)
```

```python
my_data
```

```python
#Prediction for new customer
new_predictions=knn.predict(my_data)
```
